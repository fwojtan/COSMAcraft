```python
import json
from matplotlib import pyplot as plt
import os
path_to_data = r'C:\Users\FINLA\OneDrive\DesktopHDD Docs\UNI\MISCADA\Dissertation\Usage Analysis\usage_data'
```

```python
dataset = []
```

```python
for file in os.listdir('usage_data/'):
    if 'latest' not in file:
        if '\uf03a' in file:
            new_filename = file.replace('\uf03a', '-')
            os.rename(path_to_data+'\\'+file, path_to_data+'\\'+new_filename)
            file = new_filename
            
        with open(path_to_data+'\\'+file) as f:
            data = json.load(f)
            dataset.append(data)
    
```

```python
memory_avg = 0.0
cpu_avg = 0.0
count = 0
mem_uses = []
cpu_uses = []
word_string = ''
for snapshot in dataset:
    for node in snapshot:
        cpu_avg += snapshot[node]['cpu']
        cpu_uses.append(snapshot[node]['cpu'])
        memory_avg += snapshot[node]['mem']
        mem_uses.append(snapshot[node]['mem'])
        if snapshot[node]['job'] != 'None':
            word_string += ' '+snapshot[node]['job']
        count += 1
```

```python
memory_avg /= count
cpu_avg /= count
```

```python
print(memory_avg)
print(cpu_avg)
```

```python
for i, cpu in enumerate(cpu_uses):
    if cpu>100:
        cpu_uses[i] = 100
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].hist(mem_uses, bins=50)
ax[0].set_xlabel('Memory Usage %')
ax[1].hist(cpu_uses, bins=50)
ax[1].set_xlabel('Core Usage %')
ax[0].set_ylabel('Occurences')
ax[0].set_title('COSMA 7 Memory Usage')
ax[1].set_title('COSMA 7 CPU Usage')
plt.show()
```
